## AIM: check frequencies of feature types from the statistical features selected by BORUTA

```python
import pandas as pd
import numpy as np
import pickle
```

```python
df_stat_features = pd.read_pickle(r'D:\Documents\RU\Master_Neurobiology\Internship_jaar_2\Project\TD-BRAIN\TD-BRAIN_extracted_features\df_selected_stat_features.pkl')
df_stat_features = df_stat_features.dropna(subset=['diagnosis'])
df_stat_features.sample(7)
```

```python
# create 3 feature sets [EC, EO, ratio] with column ['ID', 'age', 'gender', 'diagnosis', 'epoch']
df_stat_features_ec = df_stat_features[df_stat_features.columns[~(df_stat_features.columns.str.startswith('EO') | df_stat_features.columns.str.startswith('ratio'))]]
df_stat_features_eo = df_stat_features[df_stat_features.columns[~(df_stat_features.columns.str.startswith('EC') | df_stat_features.columns.str.startswith('ratio'))]]
df_stat_features_ratio = df_stat_features[df_stat_features.columns[~(df_stat_features.columns.str.startswith('EC') | df_stat_features.columns.str.startswith('EO'))]]
```

```python
# count the number of columns names containing 'mean', 'std', 'median', 'skew', 'kurt' for each feature set
print('EC:')
print('mean:', df_stat_features_ec.columns.str.contains('mean').sum())
print('std:', df_stat_features_ec.columns.str.contains('std').sum())
print('median:', df_stat_features_ec.columns.str.contains('median').sum())
print('skew:', df_stat_features_ec.columns.str.contains('skew').sum())
print('kurt:', df_stat_features_ec.columns.str.contains('kurt').sum())
print('')
print('EO:')
print('mean:', df_stat_features_eo.columns.str.contains('mean').sum())
print('std:', df_stat_features_eo.columns.str.contains('std').sum())
print('median:', df_stat_features_eo.columns.str.contains('median').sum())
print('skew:', df_stat_features_eo.columns.str.contains('skew').sum())
print('kurt:', df_stat_features_eo.columns.str.contains('kurt').sum())
print('')
print('Ratio:')
print('mean:', df_stat_features_ratio.columns.str.contains('mean').sum())
print('std:', df_stat_features_ratio.columns.str.contains('std').sum())
print('median:', df_stat_features_ratio.columns.str.contains('median').sum())
print('skew:', df_stat_features_ratio.columns.str.contains('skew').sum())
print('kurt:', df_stat_features_ratio.columns.str.contains('kurt').sum())
```
