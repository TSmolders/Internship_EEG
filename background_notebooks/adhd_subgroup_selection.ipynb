```python
import os
import numpy as np
import pandas as pd
import shutil
```

```python
# set folder containing participants data as working directory
working_directory = "D:\Documents\RU\Master_Neurobiology\Internship_jaar_2\Project\TD-BRAIN\TDBRAIN_participants_V2_data"
os.chdir(working_directory)
os.getcwd()
```

```python
# load participants data
df_full = pd.read_pickle('df_participants.pkl')
df_full
```

```python
# select participants with atleast an indication of ADHD or healthy
df_adhd = df_full[df_full['indication'].str.contains('ADHD|HEALTHY') == True]
df_adhd
```

```python
# count number of participants with an indication of ADHD and healthy
df_ind_adhd = df_adhd[df_adhd['indication'].str.contains('ADHD|HEALTHY') == True]
print(f"# of participants formally diagnosed with ADHD: {df_ind_adhd[df_ind_adhd['indication'].str.contains('ADHD')].shape[0]}")
print(f"# of healthy participants: {df_ind_adhd[df_ind_adhd['indication'].str.contains('HEALTHY')].shape[0]}")
```

```python
# select participants with a formal diagnosis of ADHD or healthy
df_formal_adhd = df_adhd[df_adhd['formal_status'].str.contains('ADHD|HEALTHY') == True]
print(f"# of participants formally diagnosed with ADHD: {df_formal_adhd[df_formal_adhd['formal_status'].str.contains('ADHD')].shape[0]}")
print(f"# of healthy participants: {df_formal_adhd[df_formal_adhd['formal_status'].str.contains('HEALTHY')].shape[0]}")
df_formal_adhd
```

```python
# randomly sample 20 participants from ADHD group and healthy group
n = 25
df_sample_adhd = df_formal_adhd.groupby('formal_status', group_keys=False, observed=False).apply(lambda x: x.sample(min(len(x), n), random_state=42), include_groups=True).reset_index(drop=True)
df_sample_adhd
```

```python
# create directory containing the sample raw EEG data

eeg_dir = "D:\Documents\RU\Master_Neurobiology\Internship_jaar_2\Project\TD-BRAIN\TDBRAIN-dataset-derivatives\derivatives"
exlude_dirs = ['preprocessed', 'results_manuscript'] # exclude these directories
sample_ids = df_sample_adhd['participants_ID'].tolist() # list of participants to include
sample_dir = "D:\Documents\RU\Master_Neurobiology\Internship_jaar_2\Project\TD-BRAIN\TDBRAIN-dataset-derivatives\derivatives/adhd_sample"

for subdir, dirs, files in os.walk(eeg_dir): # iterate through all files
    dirs[:] = [d for d in dirs if d not in exlude_dirs] # exclude directories
    for file in files:
        if any(sample_id in file for sample_id in sample_ids): # filter participants to include
            if 'ses-1' in file: # filter first session
                
                sample_path = os.path.join(sample_dir, file.split('_')[0]) # create directory including ID to copy files to

                path_file = os.path.join(subdir, file) # path to original file

                os.makedirs(sample_path, exist_ok=True) # create subdirectories if it does not exist

                shutil.copy2(path_file, sample_path) # copy file to sample directory
```

```python
# create directory containing the sample preprocessed EEG data

eeg_dir = "D:\Documents\RU\Master_Neurobiology\Internship_jaar_2\Project\TD-BRAIN\TDBRAIN-dataset-derivatives\derivatives\preprocessed"
# exlude_dirs = ['preprocessed', 'results_manuscript'] # exclude these directories
sample_ids = df_sample_adhd['participants_ID'].tolist() # list of participants to include
sample_dir = "D:\Documents\RU\Master_Neurobiology\Internship_jaar_2\Project\TD-BRAIN\TDBRAIN-dataset-derivatives\derivatives/adhd_sample/preprocessed"

for subdir, dirs, files in os.walk(eeg_dir): # iterate through all files
    #dirs[:] = [d for d in dirs if d not in exlude_dirs] # exclude directories
    for file in files:
        if any(sample_id in file for sample_id in sample_ids): # filter participants to include
            if 'ses-1' in file: # filter first session
                
                sample_path = os.path.join(sample_dir, file.split('_')[0]) # create directory including ID to copy files to

                path_file = os.path.join(subdir, file) # path to original file

                os.makedirs(sample_path, exist_ok=True) # create subdirectories if it does not exist

                shutil.copy2(path_file, sample_path) # copy file to sample directory
```
